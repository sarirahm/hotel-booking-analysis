# 1. Business Background

Hotel booking cancellation is one of the biggest challenges in the hospitality industry. A high cancellation rate can reduce occupancy, impact revenue forecasting, and make operational planning less efficient

This project aims to analyze hotel booking data to identify the factors associated with booking cancellations and provide data-driven recommendations for improving business performance

---

# 2. Business Problem

Hotel management has observed a high number of booking cancellations. However, the underlying causes remain unclear.

Without understanding the drivers behind cancellations, the company may face:

- Loss of potential revenue
- Difficulty managing room availability
- Inefficient staffing and operational planning
- Lower customer satisfaction

Therefore, an analysis is needed to uncover the factors influencing booking cancellations.

---

# 3. Business Objective

The objective of this project is to analyze hotel booking data and identify the main factors associated with booking cancellations.

The findings will help management develop strategies to reduce cancellation rates and improve hotel occupancy.

---

# 4. Business Questions

This project aims to answer the following questions:

1. What is the overall booking cancellation rate?
2. Does cancellation differ between Resort Hotel and City Hotel?
3. Does lead time affect cancellation?
4. Which months have the highest cancellation rate?
5. Are repeat guests less likely to cancel?
6. Which market segments contribute most to cancellations?
7. Which customer types are most likely to cancel?

---

# 5. Success Metrics

The project will evaluate several business metrics:

| KPI | Description |
|------|-------------|
| Total Bookings | Number of hotel bookings |
| Cancellation Rate | Percentage of canceled bookings |
| Average Daily Rate | Average room price |
| Occupancy Rate | Percentage of successful bookings |
| Revenue Loss | Estimated revenue lost due to cancellations |

---

# 6. Initial Hypotheses

Before analyzing the data, several hypotheses are proposed:

- Customers with longer lead time are more likely to cancel
- City Hotels have a higher cancellation rate than Resort Hotels
- Repeat guests tend to cancel less frequently
- Certain market segments have significantly higher cancellation rates
- Seasonal patterns influence booking cancellations

These hypotheses will be validated during the Exploratory Data Analysis (EDA)
