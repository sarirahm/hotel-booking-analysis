## Statistical Summary

Descriptive statistics provide an overview of the distribution of numerical variables. This helps identify unusual values, understand customer behavior, and determine whether further investigation is needed before analysis.

```python
df.describe().T
```

### Booking Behavior

Customers typically book their stay around **104 days before arrival**. However, the maximum lead time reaches **737 days**, indicating that some reservations are made nearly two years in advance. Such a wide range suggests that booking behavior varies considerably and that lead time may be an important factor influencing booking cancellations.

Most bookings involve approximately **two adults**, which aligns with typical hotel booking behavior. However, the maximum value of **55 adults** appears unusually high and should be examined further to determine whether it represents a group reservation or an anomaly.

On average, customers make **fewer than one special request** per booking, indicating that additional service requests are relatively uncommon among hotel guests.

### Potential Data Quality Issues

The ADR ranges from **-6.38 to 5400**, indicating the presence of extreme values. Since a negative room rate is generally not realistic, this variable requires further investigation during the data cleaning stage to determine whether these values represent data entry errors or valid business cases.

The `children` variable contains only **4 missing values**. Although the proportion is negligible, these observations should still be reviewed during data cleaning to ensure data consistency.

Only a small portion of bookings is associated with a company, explaining the large number of missing values in the `company` variable. Similarly, although most bookings are linked to travel agents, some records contain missing values in the `agent` variable. These variables require further assessment before selecting an appropriate handling strategy.

### Exceptional Cases

Most bookings are **not placed on a waiting list**, as indicated by the median value of zero. However, a small number of bookings experienced waiting periods of up to **391 days**, suggesting the existence of exceptional booking cases that may warrant further investigation.

### Overall Observation

Based on the descriptive statistics, several key observations can be identified:

- Booking behavior varies considerably, particularly in terms of lead time, suggesting that customers plan their trips differently.
- Several variables contain extreme or potentially unrealistic values (e.g., ADR and number of adults), which should be validated during the data cleaning stage.
- Most bookings involve two adults, with relatively few children, babies, or special service requests.
- Missing values are concentrated in the `company` and `agent` variables, indicating that not all bookings are made through companies or travel agents.
- Overall, the dataset appears sufficiently complete and suitable for further exploratory analysis after addressing data quality issues.
